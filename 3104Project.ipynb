```python
#@title
from IPython.display import HTML, display, Markdown, Video, clear_output
from ipywidgets import Layout
import ipywidgets as widgets
import os


# layout init.
btn_layout = widgets.Layout(width='45%', height='40px')
btn_sm_layout = widgets.Layout(width='20%', height='40px') 
label_layout = Layout(width='200px',height='auto')
ddl_layout = widgets.Layout(width='45%', height= 'auto')
full_layout =  widgets.Layout(width='100%', height='auto') 

def load_TSU():
    # !pip install python-dotenv -q
    # !pip install omegaconf -q
    # !pip install wandb -q

    %cd /content/drive/MyDrive/ict3104-team03-2022/

    # Functions to use along widgets
    ##General Imports
    from google.colab import output
    output.enable_custom_widget_manager()
    from ipywidgets import Layout
    # !pip install python_dotenv
    from dotenv import load_dotenv
    import wandb
    import os
    from ipywidgets import widgets
    # Faciliate file selection
    from tkinter import filedialog
    # Widget Packages
    import ipywidgets as widgets
    # jupyter nbextension enable --py widgetsnbextension
    # Used for local directory
    import os
    import sys
    # import TSU.json_util as json_util
    # from TSU.json_util import test_train_json

    from IPython.display import HTML
    from util_modules.ui_util.progress_bar import progress_bar

    ##Imports for captioning
    import pandas
    import cv2
    import re

    ##Imports for Charting
    import wandb

    ##Imports for Feature Extraction
    from os import walk
    from video_features.utils.utils import build_cfg_path
    from omegaconf import OmegaConf
    import torch
    from video_features.features_models.i3d.extract_i3d import ExtractI3D

    ##Turn off warnings for python
    import warnings
    warnings.filterwarnings('ignore')
    warnings.simplefilter('ignore')

    ##Imports for TSU
    from util_modules.json_util.test_train_json import inference_json

    print("imports executed successfully.")


    # Widget
    btn_check_if_cuda = widgets.Button(description="Check if your device is CUDA supported",
        layout = btn_layout, button_style='info')

    cuda_output = widgets.Output()

    def check_if_cuda(b):
        with cuda_output:
            cuda_output.clear_output()
            if (torch.cuda.is_available()):
                print('Your computer is CUDA supported.')
            else:
                print('Your computer is not CUDA supported.')
        # --- write script here to check if user's machine if cuda supported
        return True

    btn_check_if_cuda.on_click(check_if_cuda)

    # Display
    display(btn_check_if_cuda, cuda_output)
    
    display(Markdown("# **Toyota Smart Home Untrimmed**"))
    display(Markdown("The following Machine Learning (ML) pipelines have been integrated with this project to allow you to have a hand at Activity Detection with ease. Click on any of the following links to learn more about their source repositiories if you're interested."))
    display(Markdown("* [Toyota Smarthome (TSU)](https://project.inria.fr/toyotasmarthome/)"))
    display(Markdown("* [NVIDIA STEP)](https://github.com/NVlabs/STEP/)"))
    display(Markdown("Issue No: https://ict3104-team03-2022.atlassian.net/browse/T03-31"))
    display(Markdown("### Data Exploration"))
    display(Markdown("Select and view playback of any video from the `data` folder. "))
    display(Markdown("Issue No: https://ict3104-team03-2022.atlassian.net/browse/T03-18"))

    from base64 import b64encode
    # output.enable_custom_widget_manager()
    # Logic
    def populateList(fileDirectory, fileType=""):
        folder_files = os.listdir(fileDirectory) #You can also use full path.
        print("`/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos` directory contains {len_folder} file(s).".format(len_folder=len(folder_files)))
        fileList = []
        for file in folder_files:
            if file.endswith(fileType) or fileType == "":
                fileList.append(file)
        return fileList

    fileList = populateList('/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos', ".mp4")
            
    # Widget
    label_data_files= widgets.Label("Select Video(s):", layout = label_layout)
    ddl_data_selected = widgets.Dropdown(
        options = fileList,
        value= fileList[0],
        layout = ddl_layout)

    btn_video = widgets.Button(
        description='Play video', layout = btn_sm_layout, button_style='info'
    )

    btn_clear = widgets.Button(
        description='Clear Playback', layout = btn_sm_layout
    )

    # Display output
    video_output = widgets.Output()
    def explore_video(b):
        with video_output:
            # Create Video File
            video_output.clear_output()
            videoDir = "/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos/{}".format(ddl_data_selected.value)
            mp4 = open(videoDir,'rb').read()
            data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
            html_video_code = '<video width="80%" height="80%" controls><source src="{}" type="video/mp4"></video>'.format(data_url)
            video_output.append_display_data(HTML(html_video_code))

    btn_video.on_click(explore_video)

    # CLear output
    def clear_video(b):
        with video_output:
            video_output.clear_output()
            
    btn_clear.on_click(clear_video)

    # Display
    data_box = widgets.VBox([widgets.HBox([label_data_files, ddl_data_selected])])
    data_btn_box = widgets.VBox([widgets.HBox([btn_clear, btn_video]),video_output])
    display(data_box)
    display(data_btn_box)
    # display(widgets.HBox([label_data_files, ddl_data_selected]))
    # data_btn_box
    
    display(Markdown("***"))
    display(Markdown("### TSU Feature Extraction"))
    display(Markdown("This section performs feature extractions from videos into numpy format, so that it can be used to perform inferencing, training and testing."))
    display(Markdown("#### Parameters"))
    display(Markdown("* Folder & File"))
    display(Markdown("  * Locate the directory where your videos are stored for feature extraction"))
    display(Markdown("  * Select one one or more videos"))
    display(Markdown("* Stream Type"))
    display(Markdown("  * Select between `null, rgb, flow` stream types"))
    
    
    #### Widgets
    %cd /content/drive/MyDrive/ict3104-team03-2022/video_features
    from video_features.utils.utils import build_cfg_path
    from omegaconf import OmegaConf
    import torch
    from video_features.features_models.i3d.extract_i3d import ExtractI3D

    sub_folders = [name for name in os.listdir('/content/drive/MyDrive/ict3104-team03-2022/data') if os.path.isdir(os.path.join('/content/drive/MyDrive/ict3104-team03-2022/data', name))]
    # sub_folders.remove('.ipynb_checkpoints')
    sub_folders = sorted(sub_folders,reverse=True)

    list_of_videos = []

    label_video_folder = widgets.Label("Select Folder:", layout = label_layout)
    ddl_video_folder = widgets.Dropdown(
        options=sub_folders,
        layout = btn_layout)

    label_stream = widgets.Label("Select Stream:", layout = label_layout)
    ddl_stream = widgets.Select(
        options=['null', 'rgb', 'flow'],
        value='null',
        layout = ddl_layout)

    btn_confirm_folder = widgets.Button(description="Confirm Folder", layout = btn_layout, button_style='info')

    feature_files = {}

    def selectvideos(b):
        for (dirpath, dirnames, filenames) in walk('/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos/'):
            list_of_videos.extend(filenames)
            break
        label_feature_files= widgets.Label("Select Video(s):", layout = label_layout)
        ddl_feature_files = widgets.SelectMultiple(
            options=list_of_videos,
            layout = ddl_layout)
        ddl_feature_files.observe(on_change_videos)
        
        test_box = widgets.VBox([widgets.HBox([label_feature_files, ddl_feature_files]),
                        widgets.HBox([label_stream, ddl_stream]), btn_extract])
        display(test_box)
        


        
    #on change method for ddl_feature_files
    def on_change_videos(change):
        if change['type'] == 'change' and change['name'] == 'value':
            feature_files['name'] = list(change['new'])
    # #         feature_files = change['new'] 
    #         feature_files.append('P02T02C06.mp4')

            

    btn_confirm_folder.on_click(selectvideos)


    ### Functionality

    class Extraction:
        
        def __init__(self, feature_type):
            self.args = OmegaConf.load(build_cfg_path(feature_type))
            
        def editExtraction(self, videos_txt, stream_type):
            # self.args.show_pred = "true"
            self.args.output_path = "/content/drive/MyDrive/ict3104-team03-2022/dataExtractionOutput"
            self.args.video_paths = videos_txt
            self.args.streams = stream_type
            self.args.flow_type = "raft"
            self.args.device = "cuda:0"
            self.args.file_with_video_paths = ""
            self.args.on_extraction = "save_numpy"
            
        def showArgs(self):
            print(self.args)
            
        def runExtraction(self):
            # Load the model
            extractor = ExtractI3D(self.args)

            # Extract features
            for video_path in self.args.video_paths:
                print(f'Extracting for {video_path}')
                # save to output
                extractor._extract(video_path)
                
    # retrieve function to run extraction tool
    extracted = Extraction('i3d')

    # Function to upload own dataset from computer
    def upload_videos(file_val):
        if not file_val:
            print ("No file uploaded")

        else:   
            uploaded_video_filename =[]
            progress_bar_instance = progress_bar(len(file_val['name']))
            progress_bar_instance.display_bar()
            for item in file_val['name']: 
                uploaded_video_filename.append("/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos/"+item)
    #             print(str(uploaded_video_filename))
    #             uploaded_video_filename = item["name"]

                # save the names
                extracted.editExtraction(uploaded_video_filename, ddl_stream.value)
                extracted.runExtraction()
                print(f'{uploaded_video_filename} is extracted to "dataExtractionOutput" folder successfully!')
                progress_bar_instance.update_bar()
            try:
                
                # resets
                uploaded_file = ()
                uploaded_filename = ""
                content = ""
            except:
                print(sys.exc_info())




    def on_feature_button_clicked(b):
        upload_videos(feature_files)

    btn_extract = widgets.Button(description="Extract Features", layout = btn_layout, button_style='info')

    btn_extract.on_click(on_feature_button_clicked)  

    # Display
    feature_box = widgets.VBox([widgets.HBox([label_video_folder, ddl_video_folder]),
                                btn_confirm_folder])

    display(feature_box)
    display(Markdown("***"))
    display(Markdown("### Inference"))
    display(Markdown("This is the inference section where users can test and evaluate how accurate their model is in the actual video itself, comparing the ground truth and generated annotations."))
    display(Markdown("#### Parameters"))
    display(Markdown("* Model"))
    display(Markdown("  * Select from one of the pre-trained ML models"))
    display(Markdown("* Input Video "))
    display(Markdown("  * Select an input video from the `/data/rgbVideos` directory"))
    display(Markdown("Epic: https://ict3104-team03-2022.atlassian.net/browse/T03-33"))
    display(Markdown("Issues:"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-19"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-20"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-21"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-39"))
    
    %cd /content/drive/MyDrive/ict3104-team03-2022/

    # Logic
    model_dir = '/content/drive/MyDrive/ict3104-team03-2022/TSU/models/'
    # listdir() returns a list containing the names of the entries in the directory given by path.
    modelList = os.listdir("/content/drive/MyDrive/ict3104-team03-2022/TSU/models")
    train_modelList = os.listdir("/content/drive/MyDrive/ict3104-team03-2022/TSU/PDAN")
    modelList += train_modelList

    ## retrieve all TSU videos to dropdown list
    folder_files = os.listdir('/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos') 
    print("/content/drive/MyDrive/ict3104-team03-2022/data/exploreVideos` directory contains {len_folder} file(s).".format(len_folder=len(folder_files)))
    fileList=[]
    for file in folder_files:
        fileList.append(file)

    # Widgets
    label_inference_model = widgets.Label("Select Model:", layout = label_layout)
    ddl_inference_model = widgets.Dropdown(
        options = modelList,
        value = modelList[0],
        layout = btn_layout)

    label_inference_video = widgets.Label("Select Input Video:", layout = label_layout)
    selected_input_video = widgets.Dropdown(
        options = fileList,
        value= fileList[0],
        layout = btn_layout)

    btn_inference = widgets.Button(description="Run Inferencing", layout = btn_layout, button_style='info')

    # Display
    model_box = widgets.HBox([label_inference_model, ddl_inference_model])
    inference_box = widgets.VBox([model_box, widgets.HBox([label_inference_video, selected_input_video])])
    inference_output = widgets.Output()
    display(inference_box)
    display(btn_inference)
    display(inference_output)

    ###### Generate and Save captions Function ###################
    class Captioning:
        def __init__(self, annotation_file_path, truth_file_path, video_file):
            self.sf = pandas.read_csv(annotation_file_path, header=0, names=['action','start','end', 'accuracy'], usecols=['action','start','end','accuracy'])[['action','start','end','accuracy']]
            self.df = pandas.read_csv(truth_file_path)
            self.video_path = video_file
            self.time = 0
            self.nextTime = 0
            self.nextActionTime = 0
            self.eventCounter = 0
            self.actionCounter = 0
            self.prediction = 0.0
            self.ground_truth = ""
            self.caption = ""
            self.output_path = "/content/drive/MyDrive/ict3104-team03-2022/videosCaptionOutput/"

        def getGroundTruth(self):
            if self.time == 0:
                self.nextTime = self.df['start_frame'].iloc[1]

            if self.time >= self.nextTime:
                try:
                    self.ground_truth = self.df['event'].iloc[self.eventCounter]
                    self.ground_truth =  re.sub('[_,]', ' ', self.ground_truth)
                    self.nextTime = self.df['start_frame'].iloc[self.eventCounter+1]
                    self.eventCounter += 1
                except:
                    pass
                
        def getCaption(self):
            if self.actionCounter == 0:
                self.actionCounter, = self.sf.index[self.sf['end'] == 'end']
                
            if self.time >= self.nextActionTime:
                self.actionCounter += 1
            try:
                self.nextActionTime = float(self.sf['end'].iloc[self.actionCounter])
                self.caption = self.sf['action'].iloc[self.actionCounter]
                self.caption =  re.sub('[_,]', ' ', self.caption)
                self.prediction = self.sf['accuracy'].iloc[self.actionCounter]
            except:
                pass
                # index, = self.sf.index[self.sf['action'] == self.caption]


        def saveVideo(self):
            if not os.path.exists('/content/drive/MyDrive/ict3104-team03-2022/videosCaptionOutput/'):
                os.makedirs('/content/drive/MyDrive/ict3104-team03-2022/videosCaptionOutput/')
            cap = cv2.VideoCapture(self.video_path)
            # and our buffer to write frames
            fourcc = cv2.VideoWriter_fourcc(*'FMP4')
            out = cv2.VideoWriter(self.output_path+str(self.video_path[len(self.video_path)-13:len(self.video_path)-4])+'.mp4', fourcc, 25, (int(cap.get(3)),int(cap.get(4))))
            if (cap.isOpened() == False):
                print("Error opening video stream or file")
            counter = 0
            length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            progress_bar_instance = progress_bar(length)
            progress_bar_instance.display_bar()
            while (cap.isOpened()):
                try:
                    ret, frame = cap.read()

                    if not ret:
                        print(f'Video annotation is process complete.')
                        break

                    height, width, channels = frame.shape

                    self.time = int(cap.get(cv2.CAP_PROP_POS_MSEC)/40)
                    self.getGroundTruth()
                    self.getCaption()

                    cv2.rectangle(frame, (int(width * 0.05), int(height * 0.8)), (int(width * 0.95), int(height*0.98)), (0,0,0), -1)
                    cv2.putText(frame, "Ground truth", (int(width*0.1),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    self.arrayGroundTruth = self.ground_truth.split('.')
                    cv2.putText(frame, self.arrayGroundTruth[0], (int(width*0.1),int(height*0.9)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)
                    if len(self.arrayGroundTruth) == 2:
                        cv2.putText(frame, self.arrayGroundTruth[1], (int(width*0.1),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)

                    cv2.putText(frame, "Prediction", (int(width*0.4),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    self.arrayCaption = self.caption.split('.')
                    if self.arrayCaption[0] == self.arrayGroundTruth[0]:
                        cv2.putText(frame, self.arrayCaption[0], (int(width*0.4),int(height*0.9)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1)
                        cv2.putText(frame, self.arrayGroundTruth[0], (int(width*0.1),int(height*0.9)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1)

                    else:
                        cv2.putText(frame, self.arrayCaption[0], (int(width*0.4),int(height*0.9)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)


                    if len(self.arrayCaption) == 2:
                        if self.arrayCaption[1] == self.arrayGroundTruth[1]:
                            cv2.putText(frame, self.arrayCaption[1], (int(width*0.4),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1)
                            cv2.putText(frame, self.arrayGroundTruth[1], (int(width*0.1),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1)
                        else:
                            cv2.putText(frame, self.arrayCaption[1], (int(width*0.4),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)

                    cv2.putText(frame, "Accuracy", (int(width*0.75),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    cv2.putText(frame, str(self.prediction), (int(width*0.75),int(height*0.9)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)
                except:
                    pass
                #write our frame
                out.write(frame)
    #             cv2.imshow('frame',frame)

                key = cv2.waitKey(1)
                # define the key to
                # close the window
                if key == 'q' or key == 27:
                    break
                
                progress_bar_instance.update_bar()

            cap.release()
            out.release()
            cv2.destroyAllWindows()
        
    # Usage Captioning(annotation_file_path, truth_file_path, video_file) 
    # annotation_file_path (e.g. "./data/generatedAnnotations/PDAN_TSU_RGB_P02T01C06.csv" )
    # truth_file_path (e.g. "./data/annotations/P02/P02T01C06.csv" )
    # video_file (e.g. "./data/rgbVideos/P02T01C06.mp4")

    ###### Generate CSV After running Inferencing ###########################
    def run_inferencing(b):
        with inference_output:
            inference_output.outputs=[]
            clear_output()
            video_data = selected_input_video.value.replace(".mp4","")
            inference_json(video_data, "TSU")
            if (ddl_inference_model.value == "PDAN_TSU_RGB"):
                sub_dir = "/content/drive/MyDrive/ict3104-team03-2022/TSU/models/"
            else:
                sub_dir = "/content/drive/MyDrive/ict3104-team03-2022/TSU/PDAN/"
            inf_dir = sub_dir + ddl_inference_model.value
            inference_dataset = "inference_smarthome_CS_51.json"
            dataset_type = "TSU"
            %run /content/drive/MyDrive/ict3104-team03-2022/TSU/inferencing.py -load_model $inf_dir -video_name $selected_input_video.value -dataset $inference_dataset -dataset_type $dataset_type
            sub_folder = selected_input_video.value[:3]
            gt_annotation_directory = "/content/drive/MyDrive/ict3104-team03-2022/data/annotations/{folder}/".format(folder=sub_folder)
            generated_annotation_directory = "/content/drive/MyDrive/ict3104-team03-2022/data/generatedAnnotations/"
            annotation_file = selected_input_video.value.replace(".mp4",".csv")
            gen_annotation_file = "{loaded_model}_{vid_name}".format(loaded_model=ddl_inference_model.value, vid_name=annotation_file)
            video_directory = "/content/drive/MyDrive/ict3104-team03-2022/data/rgbVideos/{vid_name}".format(vid_name=selected_input_video.value)
            print("video is processing now..")
            test1 = Captioning(generated_annotation_directory+gen_annotation_file, gt_annotation_directory+annotation_file, video_directory)
            test1.saveVideo()
            videoDir = "/content/drive/MyDrive/ict3104-team03-2022/videosCaptionOutput/{}".format(selected_input_video.value)
            compressed_video = '/content/drive/MyDrive/ict3104-team03-2022/videosCaptionOutput/compress_video.mp4' 
            os.system(f"ffmpeg -i {videoDir} -vcodec libx264 {compressed_video}")
            mp4 = open(compressed_video,'rb').read()
            data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
            html_video_code = '<video width="80%" height="80%" controls><source src="{}" type="video/mp4"></video>'.format(data_url)
            inference_output.append_display_data(HTML(html_video_code))

    btn_inference.on_click(run_inferencing)
    display(Markdown("***"))
    display(Markdown("### Training TSU"))
    display(Markdown("This is the section whereby users will train models based on desired criterion, of which there are 3 sections:"))
    display(Markdown("#### WandB Setup"))
    display(Markdown("* It is recommended to set up Weights & Biases's Wandb.ai [https://wandb.ai/] as it is used for outputting charts"))
    display(Markdown("#### Dataset Selection"))
    display(Markdown("* Users will be able to select or upload a dataset from the folder"))
    display(Markdown("#### Initialising Training"))
    display(Markdown("* Input of custom training parameters"))
    display(Markdown("#### Model Training"))
    display(Markdown("* Initiating of model training"))
    display(Markdown("Epic: https://ict3104-team03-2022.atlassian.net/browse/T03-34"))
    display(Markdown("### WandB Setup"))
    display(Markdown("This section allows users to setup the WandB project used to output the display. "))
    display(Markdown("Additionally, the API key will be requested here."))
    display(Markdown("Your API key can be found at https://wandb.ai/authorize"))
    display(Markdown("#### Run without WandB"))
    display(Markdown('* Update the DROPDOWN to skip usage of WandB. Results for each run will be <span style="text-decoration: underline">printed out</span> instead and **not stored.**'))
    display(Markdown("#### Continuing an existing project"))
    display(Markdown("* Should you require to continue logging to a previously made project, use the previously-indicated project name."))
    #Try to load .env file

    load_dotenv(os.path.join("/content/drive/MyDrive/ict3104-team03-2022/", '.env'))

    ### Labels Declared Here
    # Enter WandB API Key
    API_label = widgets.Label('WandB API Key',
        layout=label_layout
    )

    # Initiate new Project
    desired_label = widgets.Label('Project Name',
        layout=label_layout
    )

    # WandB Username
    username_label = widgets.Label('Project Name',
        layout=label_layout
    )


    ### UserInput Declared Here
    # API Key
    if os.environ.get('WANDB_API_KEY') is not None:
        API_key_wand = widgets.Text(
        value = os.environ.get('WANDB_API_KEY'),
        disabled=False
        )
    else:
        API_key_wand = widgets.Text(
        placeholder='Type something..',
        disabled=False
        )

    # Projectname
    if os.environ.get('WANDB_PROJECTNAME') is not None:
        desired_Project_name = widgets.Text(
        value = os.environ.get('WANDB_PROJECTNAME'),
        disabled=False
        )
    else:
        desired_Project_name = widgets.Text(
        placeholder='Type something..',
        disabled=False
        )

    # Username
    if os.environ.get('WANDB_USERNAME') is not None:
        project_Username = widgets.Text(
        value = os.environ.get('WANDB_USERNAME'),
        disabled=False
        )
    else:
        project_Username = widgets.Text(
        placeholder='Type something..',
        disabled=False
        )

    ### Buttons Declared Here       
    # Create Project Button
    initiate_WandB_Project = widgets.Button(
            description='Create Project',
        button_style='info',
        layout=btn_layout,
        disabled = False
    )
    # Store API Key Button
    initiate_WandB_Project0 = widgets.Button(
        description='Store API Key',
        button_style='info',
        layout=btn_layout,
        disabled = False
    )
    # Store Username Button
    initiate_WandB_Project1 = widgets.Button(
        description='Store Username',
        button_style='info',
        layout=btn_layout,
        disabled = False
    )

    ### Outputs Declared Here
    # Stored Project Name Output
    wandbOutput = widgets.Output()
    storedProject = widgets.Label('Stored Project Name',
        layout=label_layout
    )
    # Stored API Output
    apiOutput = widgets.Output()
    storedApi = widgets.Label('Stored API Key',
        layout=label_layout
    )
    # Stored Username Output
    userOutput = widgets.Output()
    storedUser = widgets.Label('Stored Username',
        layout=label_layout
    )

    ### Checking Logic
    # Project Name
    if os.environ.get('WANDB_PROJECTNAME') is not None and os.environ.get('WANDB_PROJECTNAME') != "":
        with wandbOutput:
            print("Previous Project detected: " + os.environ.get('WANDB_PROJECTNAME'))

    # API Key
    if os.environ.get('WANDB_API_KEY') is not None and os.environ.get('WANDB_API_KEY') != "":
        with apiOutput:
            print("WANDB_API_KEY detected: " + os.environ.get('WANDB_API_KEY'))

    # Username
    if os.environ.get('WANDB_USERNAME') is not None and os.environ.get('WANDB_USERNAME') != "":
        with userOutput:
            print("WANDB_USERNAME detected: " + os.environ.get('WANDB_USERNAME'))


    ### Button Functionality Declared Here
    # Store Project API
    def store_wandB_API(b):
        with apiOutput:
            apiOutput.clear_output()
            print("Your key has been stored at")
            %env WANDB_API_KEY=$API_key_wand.value
            envfile = open(".env", "w+")
            api_key = "WANDB_API_KEY=" + API_key_wand.value
            
            if os.environ.get('WANDB_PROJECTNAME') is not None or os.environ.get('WANDB_USERNAME') is not None:
                api_key = [api_key+"\n",
                        "WANDB_PROJECTNAME="+str(os.environ.get('WANDB_PROJECTNAME'))+"\n",
                        "WANDB_USERNAME="+str(os.environ.get('WANDB_USERNAME'))]
                envfile.writelines(api_key)
            else:
                envfile.write(api_key)
            envfile.close()

    # Store Username
    def store_wandB_USERNAME(b):
        with userOutput:
            userOutput.clear_output()
            print("Your username has been stored at")
            %env WANDB_USERNAME=$project_Username.value
            envfile = open(".env", "w+")
            wandB_username = "WANDB_USERNAME=" + project_Username.value
            
            if os.environ.get('WANDB_PROJECTNAME') is not None or os.environ.get('WANDB_API_KEY') is not None:
                wandB_username = [wandB_username+"\n",
                                "WANDB_PROJECTNAME="+str(os.environ.get('WANDB_PROJECTNAME'))+"\n",
                                "WANDB_API_KEY="+str(os.environ.get('WANDB_API_KEY'))]
                envfile.writelines(wandB_username)
            else:
                envfile.write(wandB_username)
            envfile.close()
            
    # Create Project Functionality
    def create_WandB(b):
        with wandbOutput:
            #Silence Weights&Biases info messages
            %env WANDB_SILENT=TRUE
            %env WANDB_PROJECTNAME=$desired_Project_name.value
            wandbOutput.clear_output()
            envfile = open(".env", "w+")
            project_name = "WANDB_PROJECTNAME=" + desired_Project_name.value
            
            if os.environ.get('WANDB_API_KEY') is not None or os.environ.get('WANDB_USERNAME') is not None:
                project_name = [project_name+"\n",
                                "WANDB_API_KEY="+str(os.environ.get('WANDB_API_KEY'))+"\n",
                                "WANDB_USERNAME="+str(os.environ.get('WANDB_USERNAME'))]
                envfile.writelines(project_name)
            else:
                envfile.write(project_name)
            envfile.close()
            print("Your project is:"+desired_Project_name.value)

    #Link the buttons
    initiate_WandB_Project.on_click(create_WandB)
    initiate_WandB_Project0.on_click(store_wandB_API)
    initiate_WandB_Project1.on_click(store_wandB_USERNAME)

    bor = 'OFFLINE'
    drop_down = widgets.Dropdown(options=['OFFLINE', 'ACTIVE'],
                                    description='',
                                    disabled=False)
    WandbObserve = widgets.Output()

    def dropdown_handler(change):
        global bor
        bor = change.new
        with WandbObserve:
            WandbObserve.clear_output()
            view_charts.disabled=False
            display(widgets.HBox([API_label,API_key_wand]))
            display(widgets.HBox([storedApi,apiOutput]))
            display(initiate_WandB_Project0)
            display(widgets.HBox([desired_label,desired_Project_name]))
            display(widgets.HBox([storedProject,wandbOutput]))
            display(initiate_WandB_Project)
            display(widgets.HBox([username_label,project_Username]))
            display(widgets.HBox([storedUser,userOutput]))
            display(initiate_WandB_Project1)

    # drop_down.observe(dropdown_handler, names='value')
    wandb_status = widgets.Label('WandB Status',layout=label_layout)
    display(widgets.HBox([wandb_status,drop_down]))


    wandOutput = widgets.Output()

    #WANDB CHART LOGIC HERE
    def viewWand(b):
        with wandOutput:
            clear_output()
            if(desired_Project_name.value is None or desired_Project_name.value == ""):
                print("Error: Missing Project Name")
            elif(project_Username.value is None or project_Username.value == ""):
                print("Error: Missing Username")
            elif(API_key_wand.value is None or API_key_wand.value == ""):
                print("Error: Missing API Key")
            else:
                %wandb $project_Username.value/$desired_Project_name.value
    view_charts = widgets.Button(
        description='View Charts',
        button_style='info',
        layout=btn_layout,
        disabled = False
    )
    view_charts.on_click(viewWand)

    display(WandbObserve)
    with WandbObserve:
        display(widgets.HBox([API_label,API_key_wand]))
        display(widgets.HBox([storedApi,apiOutput]))
        display(initiate_WandB_Project0)
        display(widgets.HBox([desired_label,desired_Project_name]))
        display(widgets.HBox([storedProject,wandbOutput]))
        display(initiate_WandB_Project)
        display(widgets.HBox([username_label,project_Username]))
        display(widgets.HBox([storedUser,userOutput]))
        display(initiate_WandB_Project1)
    display(Markdown("***"))
    display(Markdown("### Selecting Dataset Code"))
    display(Markdown("Issue: https://ict3104-team03-2022.atlassian.net/browse/T03-22"))
    %cd /content/drive/MyDrive/ict3104-team03-2022/
    from util_modules.json_util.test_train_json import create_subset_json
    from util_modules.json_util.test_train_json import count_train_test

    # Training Dataset Upload Label
    training_dataset_label = widgets.Label("Select dataset or upload your own")

    # Create list for use with user dataset selection
    datasetList = populateList("/content/drive/MyDrive/ict3104-team03-2022/TSU/tsu_data", ".json") # populate list with files from dataset directory

    if not datasetList:
        print ("No existing datasets found")

    # Allow user to select dataset
    selectedDataset = widgets.Dropdown(
        options = datasetList,
        value= datasetList[0],
        disabled=False,
    )


    def on_change(change):
        with dataset_ddl_output:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                train, test = count_train_test(str(change['new']), "TSU")
                print("Selected: %s " % change['new'])
                print("Total Number of train: %s" % train )
                print("Total Number of test: %s" % test )

    selectedDataset.observe(on_change)

    # Allow user to upload own dataset
    training_upload_dataset = widgets.Button(
        description='Upload dataset',
        layout=btn_layout,
        button_style='info',
        disabled = False
    )

    # Allow user to cancel upload
    training_upload_cancel = widgets.Button(
        description='Cancel',
        disabled = False
    )
    # Cancel upload functionality
    def upload_dataset_cancel(b):
        with training_upload_output:
            clear_output()
            
    # Link cancel button        
    training_upload_cancel.on_click(upload_dataset_cancel)

    # Output for dataset upload
    training_upload_output = widgets.Output()
    dataset_ddl_output = widgets.Output()

    # Function to upload own dataset from computer
    def upload_dataset(file_val):
        if not file_val:
            print ("No file uploaded")

        else:   
            clear_output(wait=True)
            uploaded_file = next(iter(file_val))
            uploaded_filename = uploaded_file["name"]
            content = uploaded_file["content"]
            print("File name: " + uploaded_filename)

            try:
                save_path = '/content/drive/MyDrive/ict3104-team03-2022/TSU/tsu_data/'
                completeName = os.path.join(save_path, uploaded_filename) 
                with open(completeName, 'wb') as f: f.write(content)
                print (uploaded_filename + " uploaded to 'datasets' successfully!")
                
                # resets
                uploaded_file = ()
                uploaded_filename = ""
                content = ""
            except:
                print(sys.exc_info())
                display(training_upload_cancel)

    # Upload functionality
    def show_upload(b):
        with training_upload_output:
            clear_output()
            uploader = widgets.FileUpload(accept='.json',  # Currently only accepts .json
                                        multiple=False,
                                        description = "Browse",
                                        _counter = 0
            )
            upload_button = widgets.Button(
                description='Confirm Upload',
                disabled = False
            )

            def on_button_clicked(b):
                upload_dataset(uploader.value)

            upload_button.on_click(on_button_clicked)
            display(widgets.HBox([uploader, upload_button,training_upload_cancel]))

    #Link the buttons
    training_upload_dataset.on_click(show_upload)


    style = {'description_width': 'initial'}

    # Provide user with batch size
    training_no = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        style=style,
        disabled=False
    )



    # Provide user with batch size
    testing_no = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        style=style,
        disabled=False
    )

    ##listen to changing values
    testing_output = widgets.Output()
    def on_change_testing(change):
        with testing_output:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                total = training_no.value + testing_no.value
                train = round((training_no.value/total * 100),2)
                test = round((testing_no.value/total * 100),2)
                print("Percentage of training: %s" % train)
                print("Percentage of testing: %s" %  test)

    testing_no.observe(on_change_testing)

    # Confirm Button to provide user with virtual commit
    dataset_Confirm_Button = widgets.Button(
        description='Confirm',
        layout=btn_layout,
        button_style='info',
        disabled = False
    )
    # Output to contain printout for ALL user input
    datasetOutput = widgets.Output()


    training_UserInput_label = widgets.Label("Current User Input")
    # Functionality for user input
    def on_button_click(b):
        with datasetOutput:
            json_output = create_subset_json(testing_no.value, training_no.value, selectedDataset.value, "/content/drive/MyDrive/ict3104-team03-2022/TSU/tsu_data/", "train_" +selectedDataset.value, "/content/drive/MyDrive/ict3104-team03-2022/TSU/tsu_data/")
            '''Print out the user input'''
            datasetOutput.clear_output() # Clear previous output
            print("Total train and test video data for training: ", json_output)

    dataset_Confirm_Button.on_click(on_button_click)

    # Labels
    training_dataset_label = widgets.Label("Dataset", layout=label_layout)
    training_no_label = widgets.Label("No. of Training", layout=label_layout)
    training_testing_no_label = widgets.Label("No. of Testing", layout=label_layout)

    # User Interface
    uploadBox = widgets.HBox([training_upload_dataset,training_upload_output])
    dataset_input_box = widgets.VBox([widgets.HBox([training_dataset_label,selectedDataset, dataset_ddl_output]),
                                    uploadBox,
                                    widgets.HBox([training_no_label,training_no, testing_output]),
                                    widgets.HBox([training_testing_no_label,testing_no]),datasetOutput])
    display(dataset_input_box)
    display(dataset_Confirm_Button)
    display(Markdown("***"))
    display(Markdown("### Initialise Training"))
    display(Markdown("| Variable | Description |"))
    display(Markdown("| :- | :- |"))
    display(Markdown("| Video Type | Select from the various types (However only RGB is supported for now) |"))
    display(Markdown("|Model Name|Name of the model|"))
    display(Markdown("|Batch Size||"))
    display(Markdown("|Epoch|Number of training iterations|"))
    display(Markdown("|Kernel|Number of Concurrent Processes|"))
    display(Markdown("Issues: "))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-23"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-24"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-25"))

    class user_input:
        def __init__(self):
            self.model_name = None
            self.batch_size = None
            self.epoch = None
            self.kernelsize = None
            self.mode = 'RGB'
            self.comp_info = 'TSU_CS_RGB_PDAN'
            self.train = True
            self.dataset_type = 'TSU'
            self.dataset = None
            self.num_channel = 512
            self.APtype = 'map'
            self.model = 'PDAN'
            self.lr = 0.0002

            
    modes = ['rgb','skeleton']
    # Allow user to select type of TSU evaluation
    training_video_mode = widgets.Dropdown(
        options = modes,
        value= modes[0],
        disabled=False,
    )
            

    # Specify a name for this new model using appropriate UI elements.
    desired_model_name = widgets.Text(
        placeholder='Type something..',
        disabled=False
    )

    # Create list for use with user dataset selection
    datasetList = populateList("/content/drive/MyDrive/ict3104-team03-2022/TSU/models","") # populate list with files from dataset directory

    if not datasetList:
        print ("No existing datasets found")


    # Provide user with batch size
    user_batch_size = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        disabled=False
    )

    # Provide user with epoch selection
    epochs = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        disabled=False
    )

    # Provide user with kernel input
    kernel = widgets.BoundedIntText(
        min=2,
        # max=1000, # to be defined and added as necessary
        step=1,
        disabled=False
    )

    # Confirm Button to provide user with virtual commit
    training_Confirm_Button = widgets.Button(
        description='Confirm Input',
        layout=btn_layout,
        button_style='info',
        disabled = False
    )

    # Output to contain printout for ALL user input
    trainingParamOutput = widgets.Output()
    with trainingParamOutput:
        print("Currently Empty")

    user = user_input() #Instantiate class here
    training_UserInput_label = widgets.Label("Currently Stored Input")
    # Functionality for user input
    def on_button_click(b):
        with trainingParamOutput:
            '''Store the user input into variables'''
            train_data = "train_" + selectedDataset.value
            user.dataset = train_data
            userModelName = desired_model_name.value
            userModelName = userModelName.replace(" ", "_") # strip spaces
            user.model_name = userModelName
            user.batch_size = user_batch_size.value
            user.epoch = epochs.value
            user.kernelsize = kernel.value
            user.mode = training_video_mode.value
            '''Print out the user input'''
            trainingParamOutput.clear_output() # Clear previous output
            print(train_data)
            for item in vars(user): # Clear previous output
                print("{}:{}".format(item,vars(user)[item]))

    training_Confirm_Button.on_click(on_button_click)
    # Labels
    training_video_label = widgets.Label("Video Type", layout=label_layout)
    training_modelName_label = widgets.Label("Model Name", layout=label_layout)
    training_batchSize_label = widgets.Label("Batch Size", layout=label_layout)
    training_epochs_label = widgets.Label("Epoch", layout=label_layout)
    training_kernel_label = widgets.Label("Kernel", layout=label_layout)

    inputBox = widgets.VBox([widgets.VBox([widgets.HBox([training_video_label, training_video_mode]),
    widgets.HBox([training_modelName_label, desired_model_name]),
    widgets.HBox([training_batchSize_label, user_batch_size]),
    widgets.HBox([training_epochs_label, epochs]),
    widgets.HBox([training_kernel_label, kernel])]),widgets.VBox([training_UserInput_label,trainingParamOutput])])
    display(inputBox)

    # Select Dataset
    split_setting = ""
    if "CS" in selectedDataset.value:
        split_setting = "CS"
    elif "CV" in selectedDataset.value:
        split_setting = "CV"

    bigRedButton = widgets.Button(
        description='Train Model',
        disabled = False,
        button_style='info',
        layout=btn_layout
    )

    # Output train
    trainOutput = widgets.Output()

    # Functionality for user input
    def runTrain(b):
        with trainOutput:
            clear_output()
            if user.model_name == "" or user.model_name == None:
                print("Model name required.")
            elif desired_Project_name.value == "" or desired_Project_name.value == None:
                print("Project name required.")
            elif os.environ.get('WANDB_API_KEY') == None:
                print("WandB API Key required.")
            else:
                %run /content/drive/MyDrive/ict3104-team03-2022/TSU/train.py -batch_size $user.batch_size \
                -model_name $user.model_name -epoch $user.epoch  \
                -kernelsize $user.kernelsize \
                -mode $user.mode -comp_info $user.comp_info -train $user.train \
                -dataset_type $user.dataset_type -num_channel $user.num_channel -APtype $user.APtype\
                -model $user.model -lr $user.lr -dataset $user.dataset -split_setting $split_setting \
                -wandb_project $desired_Project_name.value -wandb $drop_down.value

    bigRedButton.on_click(runTrain)
    display(widgets.VBox([training_Confirm_Button,bigRedButton,trainOutput]))
    display(Markdown("***"))
    display(Markdown("### Testing"))
    display(Markdown("This testing section will take the trained models that was previously generated in the training section and evaluate their results to see how accurate each model is."))
    display(Markdown("Your dataset has already been selected in section 5."))
    display(Markdown("#### Parameters"))
    display(Markdown("1. -load_model, All the trained model that was generated in the previous training"))
    display(Markdown("2. -epoch"))
    display(Markdown("3. -dataset e.g. (train_smarthome_cs.json)"))
    display(Markdown("Epic: https://ict3104-team03-2022.atlassian.net/browse/T03-35"))
    display(Markdown("Issue No:"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-26"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-27"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-28"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-29"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-30"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-55"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-56"))
    display(Markdown("- https://ict3104-team03-2022.atlassian.net/browse/T03-57"))

    # Training Dataset Upload Label
    training_dataset_label = widgets.Label("Select dataset or upload your own")

    # Create list for use with user dataset selection
    model_list = populateList("/content/drive/MyDrive/ict3104-team03-2022/TSU/PDAN/",) # populate list with files from dataset directory

    if not model_list:
        print ("No existing datasets found")

    # Allow user to select dataset
    selected_model = widgets.Dropdown(
        options = model_list,
        value= model_list[0],
        disabled=False,
    )


    def on_change(change):
        with model_ddl_output:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                print("Selected: %s " % change['new'])

    selected_model.observe(on_change)

    def run_test(b):
        with test_output:
            clear_output()
            model_w_path = "/content/drive/MyDrive/ict3104-team03-2022/TSU/PDAN/" + selected_model.value
            %run /content/drive/MyDrive/ict3104-team03-2022/TSU/test.py -load_model $model_w_path -dataset $user.dataset 

    # Widget
    btn_testing = widgets.Button(description="Run Testing", layout = btn_layout, button_style='info')

    btn_testing.on_click(run_test)

    ##output widget
    test_output = widgets.Output()
    model_ddl_output = widgets.Output()

    # Display

    display(widgets.VBox([selected_model,model_ddl_output,btn_testing,test_output]))
    display(Markdown("***"))
    display(Markdown("### View WandB Charts"))
    display(Markdown("To view the charts produced by within WandB, ensure the project's privacy settings are set to PUBLIC"))
    display(Markdown("URL Format: https://wandb.ai/username/project/overview"))
    display(Markdown("TSU-project/overview"))
    display(view_charts)
    display(wandOutput)
    display(Markdown("***"))
    
    return
    
def load_MSTCT():
    !pip install pytorch==1.9.0 -q
    # !pip install scikit-learn -q
    # !pip install numpy -q
    # !pip install h5py -q
    # !pip install opencv-python -q
    # !pip install tqdm -q
    !pip install torch==1.9.0+cu102 torchvision==0.10.0+cu102 torchaudio==0.9.0 -f https://download.pytorch.org/whl/torch_stable.html -q
    # !pip install timm==0.4.12
    # imports for UI widgets

    import pandas
    import cv2
    import re
    import os
    import json
    import csv
    import math
    import numpy as np

    class CaptioningMST:
        def __init__(self, caption_filepath = None):
            self.annotation_path = ['/content/drive/MyDrive/MSTCT/MSTCT/annotations/Charades_v1_train.csv', './annotations/Charades_v1_test.csv']    
            self.videopaths = []
            self.video_type = ''
            self.time = 0
            self.output_path = "/content/drive/MyDrive/MSTCT/MSTCT/videosCaptionOutput/"
            self.classes = []
            self.dataset = {}
            self.feature = {}
            self.caption_path = caption_filepath
            self.accuracydata = {}
        
        def get_classes(self):
            with open('/content/drive/MyDrive/MSTCT/MSTCT/annotations/Charades_v1_classes.txt', newline='') as classfile:
                lines = classfile.readlines()
                self.classes.append(lines)

        def get_video_names(self):
            with open('/content/drive/MyDrive/MSTCT/MSTCT/annotations/json_data.json', 'r') as f:
                self.accuracydata = json.load(f)
            self.videopaths = self.accuracydata.keys()

        def get_feature(self):
            for k,v in self.dataset.items():
                listOfFeatures = []
                listBeforeConcat = v[9].split(';')
                for item in listBeforeConcat:
                    listOfFeatures.append(item.split(' '))
                self.feature[k]=listOfFeatures

        def getGroundTruth(self):
            self.get_video_names()

            with open(self.annotation_path[0], newline='') as csvfile:
                spamreader = csv.reader(csvfile)
                for row in spamreader:
                    for item in self.videopaths:
                        if(row[0] == item):
                            self.dataset[row[0]]=row
                            self.video_type = 'test'

            if(not self.dataset):
                with open(self.annotation_path[1], newline='') as csvfile:
                    spamreader = csv.reader(csvfile)
                    for row in spamreader:
                        for item in self.videopaths:
                            if(row[0] == item):
                                self.dataset[row[0]]=row
                                self.video_type = 'train'
            
            if(not self.dataset): return print('Video not found in dataset') 

            self.get_classes()
            self.get_feature()
            self.createVideo()

        def createVideo(self):
            if not os.path.exists('/content/drive/MyDrive/MSTCT/MSTCT/videosCaptionOutput/'):
                os.makedirs('/content/drive/MyDrive/MSTCT/MSTCT/videosCaptionOutput/')
            for video_name in self.videopaths:
                cap = cv2.VideoCapture('/content/drive/MyDrive/MSTCT/MSTCT/data/rgbVideos/'+video_name+'.mp4')

                fourcc = cv2.VideoWriter_fourcc(*'mp4v')
                out = cv2.VideoWriter(self.output_path+video_name+'.mp4', fourcc, 25, (int(cap.get(3)),int(cap.get(4))))
                if (cap.isOpened() == False):
                    print("Error opening video stream or file")
                    return

                while (cap.isOpened()):
                    ret, frame = cap.read()

                    if not ret:
                        print(f'Video annotation for {video_name} is process complete.')
                        break
                    
                    height, width, channels = frame.shape
                    cv2.rectangle(frame, (0, int(height * 0.8)), (int(width), int(height*0.98)), (0,0,0), -1)
                    cv2.putText(frame, "Ground truth", (int(width*0.1),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    cv2.putText(frame, "Accuracy", (int(width*0.75),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    cv2.putText(frame, "Prediction", (int(width*0.4),int(height*0.85)), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1)
                    
                    self.time = int(cap.get(cv2.CAP_PROP_POS_MSEC))

                    listofGroundTruths = []
                    for key, value in self.feature.items():
                        if(key == video_name):
                            for item in value:
                                if(self.time > float(item[1])*1000 and self.time < float(item[2])*1000):
                                    listofGroundTruths.append(item[0][1:])
                    
                    feature_split = np.array_split(listofGroundTruths,2)
                    cv2.putText(frame, ', '.join(feature_split[0]), (int(width*0.1),int(height*0.90)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)
                    cv2.putText(frame, ', '.join(feature_split[1]), (int(width*0.1),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)

                    # if want to see the full text
                    # for index, item in enumerate(listofGroundTruths):
                    #     cv2.putText(frame, item, (int(width*0.2),int(height*(0.30+(index*0.05)))), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,0), 1)
                    #     cv2.putText(frame, self.classes[0][int(item)], (int(width*0.22),int(height*(0.30+(index*0.05)))), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,0), 1)
                        
                    listOfActivities = []
                    accuraciesOfActivities = []
                    for item in self.accuracydata[video_name]['actions']:
                        if(self.time > item[1]*1000 and self.time < item[2]*1000):
                            listOfActivities.append(item[0])
                            accuraciesOfActivities.append(item[3])

                    activities_split = np.array_split(listOfActivities, 2)
                    prediction_split = np.array_split(accuraciesOfActivities, 2)
                    
                    if(str(feature_split[0])==str(activities_split[0])):
                        cv2.putText(frame, ', '.join([str(item) for item in activities_split[0]]), (int(width*0.4),int(height*0.9)), cv2.FONT_HERSHEY_DUPLEX, 0.4, (67,181,75), 1)
                    else:
                        cv2.putText(frame, ', '.join([str(item) for item in activities_split[0]]), (int(width*0.4),int(height*0.9)), cv2.FONT_HERSHEY_DUPLEX, 0.4, (255,255,255), 1)
                    
                    if(str(feature_split[1])==str(activities_split[1])):   
                        cv2.putText(frame, ', '.join([str(item) for item in activities_split[1]]), (int(width*0.4),int(height*0.95)), cv2.FONT_HERSHEY_DUPLEX, 0.4, (67,181,75), 1)
                    else:
                        cv2.putText(frame, ', '.join([str(item) for item in activities_split[1]]), (int(width*0.4),int(height*0.95)), cv2.FONT_HERSHEY_DUPLEX, 0.4, (255,255,255), 1)
                        
                    cv2.putText(frame, ', '.join([str(item) for item in prediction_split[0]]), (int(width*0.75),int(height*0.90)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)
                    cv2.putText(frame, ', '.join([str(item) for item in prediction_split[1]]), (int(width*0.75),int(height*0.95)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255,255,255), 1)
                    
                    
                    out.write(frame)
                    # cv2.imshow('frame',frame)
                    key = cv2.waitKey(1)
                    # define the key to
                    # close the window
                    if key == 'q' or key == 27:
                        break

                cap.release()
                out.release()
                cv2.destroyAllWindows()
    
    display(Markdown("# ***MSTCT + Pytorch-I3D***"))
    display(Markdown("This Colab will contain the code for the 2nd algorithm, to do activity detection for charades. <br>"))
    display(Markdown("The feature extraction based on piergiaj/pytorch-i3d will also be implemented."))
    display(Markdown("*** "))
    display(Markdown("##**Feature Extraction**"))
    display(Markdown("This section performs feature extraction from jpeg folders of videos via [pytorch-i3d](https://github.com/piergiaj/pytorch-i3d)."))
    display(Markdown("#### Run the feature extraction file"))
    display(Markdown("* Add the videos jpeg folder you want to extract into `/content/drive/MyDrive/MST/pytorch-i3d/Charades_v1_rgb`"))
    display(Markdown("* Select `rgb` stream and `rgb_charades`"))
    display(Markdown("* Click on `Extact Features` button"))
    display(Markdown("* Retrieve the numpy files from `/content/drive/MyDrive/MST/pytorch-i3d/output`"))
    # to be replaced with files with /models/ later
    models_list = ['flow_charades','flow_imagenet','rgb_charades','rgb_imagenet']

    label_stream = widgets.Label("Select Stream:", layout = label_layout)
    ddl_stream = widgets.Select(
        options=['rgb', 'flow', 'both'],
        value='rgb',
        layout = ddl_layout)

    label_models = widgets.Label("Select model:", layout = label_layout)
    ddl_models = widgets.Dropdown(
        options= models_list,
        value='rgb_charades',
        layout = ddl_layout)

    btn_extract = widgets.Button(description="Extract Features", layout = btn_layout, button_style='info')

    extraction_output = widgets.Output()

    def extract_feature(b):
        with extraction_output:
            extraction_output.clear_output()
            print('run extraction ...')
            %run /content/drive/MyDrive/MSTCT/pytorch-i3d/extract_features.py -mode $ddl_stream.value -gpu "0" -root "/content/drive/MyDrive/MSTCT/pytorch-i3d/Charades_v1_rgb" -split_file "/content/drive/MyDrive/MSTCT/pytorch-i3d/charades/full_charades.json" -save_dir "/content/drive/MyDrive/MSTCT/pytorch-i3d/output" -load_model "/content/drive/MyDrive/MSTCT/pytorch-i3d/models/"+str($ddl_models.value)
    
    btn_extract.on_click(extract_feature)

    # Display
    feature_extraction_box = widgets.VBox([widgets.HBox([label_stream, ddl_stream]), widgets.HBox([label_models, ddl_models]), btn_extract, extraction_output])
    
    display(feature_extraction_box)
    display(Markdown("##**MSTCT Training**"))
    display(Markdown("#### Run the training instead of shell script"))
    display(Markdown("* Configure the training options using the widgets"))
    display(Markdown("* Rename the `pickle output file` to be used later"))
    display(Markdown("* Click on `Train Charades` to run training model"))
    display(Markdown("* Pickle file is saved in the `/content/drive/MyDrive/MST/MST/save_logit/`"))

    # to be replaced with files with /models/ later
    models_list = ['MS_TCT']

    # # select training t/f
    # label_train = widgets.Label("Run training", layout = label_layout)
    # ddl_train = widgets.Select(
    #     options=['True','False'],
    #     value='True',
    #     layout = ddl_layout)

    # select model
    label_models = widgets.Label("Select model:", layout = label_layout)
    ddl_models_training = widgets.Dropdown(
        options= models_list,
        value = "MS_TCT",
        layout = ddl_layout)

    # select stream
    label_stream = widgets.Label("Select Stream:", layout = label_layout)
    ddl_stream_training = widgets.Select(
        options=['rgb', 'flow'],
        value='rgb',
        layout = ddl_layout)


    # select batch size
    label_batch_size = widgets.Label("batch_size:", layout = label_layout)
    batch_size_training = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        value=32,
        disabled=False
    )

    # select epoch
    label_epochs = widgets.Label("epoch:", layout = label_layout)
    epochs_training = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        value=50,
        disabled=False
    )

    # select unisize
    label_unisize= widgets.Label("unisize:", layout = label_layout)
    ddl_unisize = widgets.Select(
        options=[True,False],
        value=True,
        layout = ddl_layout)

    # select alpha_1
    label_alpha= widgets.Label("alpha_1:", layout = label_layout)
    text_alpha = widgets.BoundedIntText(
        placeholder='Specify alpha_1 value',
        value = 1,
        disabled=False
    )

    # select beta_1
    label_beta= widgets.Label("beta_1:", layout = label_layout)
    text_beta = widgets.BoundedFloatText(
        placeholder='Specify beta_1 value',
        value = 0.05,
        disabled=False
    )

    # select comp_info
    label_comp_info= widgets.Label("comp_info:", layout = label_layout)
    ddl_comp_info = widgets.Select(
        options=[True,False],
        value=True,
        layout = ddl_layout)

    # select skip
    label_skip= widgets.Label("skip:", layout = label_layout)
    ddl_skip = widgets.Select(
        options=[0,1],
        value=0,
        layout = ddl_layout)

    # select lr
    label_lr= widgets.Label("lr:", layout = label_layout)
    lr_text_lr = widgets.BoundedFloatText(
        placeholder='Specify lr value',
        value=0.0001,
        disabled=False
    )

    label_save_file = widgets.Label("file name (output):", layout = label_layout)
    text_save_file = widgets.Text(
        placeholder='Specify file name',
        disabled=False
    )

    training_Confirm_Button  = widgets.Button(description="Train Charades", 
                                              layout = btn_layout, 
                                              button_style='info')

    mstct_training_output = widgets.Output()


    def train_mstct_features(b):
        with mstct_training_output:
            mstct_training_output.clear_output()
            if text_save_file.value == '':
              print('Make a unique file name')
              return
            print('run training ...')
            %run /content/drive/MyDrive/MSTCT/MS-TCT/train.py -dataset "charades" -gpu 0 -mode $ddl_stream_training.value -model $ddl_models_training.value -train True -num_clips 256 -skip $ddl_skip.value -lr $lr_text_lr.value -comp_info $ddl_comp_info.value -epoch $epochs_training.value -unisize $ddl_unisize.value -alpha_l text_alpha.value -beta_l float(text_beta.value) -batch_size $batch_size_training.value -save_file_name $text_save_file.value
            CaptioningMST().getGroundTruth()
            
    training_Confirm_Button.on_click(train_mstct_features)


    # Display
    training_mstct_box = widgets.VBox([widgets.HBox([label_models, ddl_models_training]),
                                widgets.HBox([label_stream, ddl_stream_training]), 
                                widgets.HBox([label_batch_size, batch_size_training]), 
                                widgets.HBox([label_epochs, epochs_training]), 
                                widgets.HBox([label_unisize, ddl_unisize]), 
                                widgets.HBox([label_alpha, text_alpha]), 
                                widgets.HBox([label_beta, text_beta]), 
                                widgets.HBox([label_comp_info, ddl_comp_info]), 
                                widgets.HBox([label_skip, ddl_skip]), 
                                widgets.HBox([label_lr, lr_text_lr]), 
                                widgets.HBox([label_save_file, text_save_file]), 
                                training_Confirm_Button,
                                mstct_training_output])
    
    display(training_mstct_box)
    display(Markdown("***"))
    display(Markdown("##**Testing for MSTCT**"))
    display(Markdown("* Select Pickle file to test using the drop down selection list referencing the `/content/drive/MyDrive/MST/MST/save_logit/` directory"))
    display(Markdown("* Run testing and see the probability of each video"))
    display(Markdown("* Put videos into `/content/drive/MyDrive/MST/MST/data/rgbVideos`"))
    display(Markdown("* Accuracies of each batch of video will be stored into `/content/drive/MyDrive/MST/MST/annotations/json_data.json`"))
    display(Markdown("* Wait for videos to generate to `/content/drive/MyDrive/MST/MST/videosCaptionOutput`"))
    
    def populateList(fileDirectory, fileType):
        folder_files = os.listdir("/content/drive/MyDrive/MSTCT/MS-TCT/save_logit") #You can also use full path.
        print("This Folder contains {len_folder} file(s).".format(len_folder=len(folder_files)))
        fileList = []
        for file in folder_files:
            if file.endswith(fileType):
                fileList.append(file)
        return fileList

    pickleList = populateList('/content/drive/MyDrive/MSTCT/MS-TCT/save_logit', ".pkl")
            
    ## dropdown list from the ./save_logit/ to allow user to select the output files
    label_pickleList = widgets.Label("Select Pickle file:", layout = label_layout)
    ddl_pickleList = widgets.Dropdown(
        options= pickleList,
        value= pickleList[0],
        layout = ddl_layout)

    button_mstct_execute = widgets.Button(description="Generate Video", 
                                              layout = btn_layout, 
                                              button_style='info')


    generate_video_output = widgets.Output()

    # To add onclick function here
    def execute(b):
        with generate_video_output:
          generate_video_output.clear_output()
          %cd /content/drive/MyDrive/MSTCT/MS-TCT/Evaluation.py -pkl_path /content/drive/MyDrive/MSTCT/MS-TCT/save_logit/$ddl_pickleList.value
            
    button_mstct_execute.on_click(execute)

    # Display
    testing_mstct_box = widgets.VBox([widgets.HBox([label_pickleList, ddl_pickleList]),
                                button_mstct_execute, generate_video_output])
    display(testing_mstct_box)

    display(Markdown("***"))
    display(Markdown("## Inferencing Play Video"))
    display(Markdown("* Select video to play"))
    display(Markdown("* Play video"))

    import glob
    from base64 import b64encode
    from IPython.display import HTML
    ## UI to run inferencing

    def populateList(fileDirectory, fileType=""):
        folder_files = os.listdir(fileDirectory) #You can also use full path.
        print("`/content/drive/MyDrive/MSTCT/MS-TCT/videoCaptionOutput` directory contains {len_folder} file(s).".format(len_folder=len(folder_files)))
        fileList = []
        for file in folder_files:
            if file.endswith(fileType) or fileType == "":
                fileList.append(file)
        return fileList

    fileListMST = populateList('/content/drive/MyDrive/MSTCT/MS-TCT/videosCaptionOutput', ".mp4")
            
    # Widget
    label_data_files= widgets.Label("Select Video(s):", layout = label_layout)
    ddl_data_selected = widgets.Dropdown(
        options = fileListMST,
        value= fileListMST[0],
        layout = ddl_layout)

    btn_play = widgets.Button(
        description='Play video', layout = btn_sm_layout, button_style='info'
    )

    video_output = widgets.Output()
    def run_video(b):
        with video_output:
            video_output.clear_output()
            print("Generating video...")
            # video_file = "videosCaptionOutput/" + ddl_data_selected.value
            videoDir = "/content/drive/MyDrive/MSTCT/MS-TCT/videosCaptionOutput/{}".format(ddl_data_selected.value)
            compressed_video = './compress_video.mp4' 
            os.system(f"ffmpeg -i {videoDir} -vcodec libx264 {compressed_video}")
            mp4 = open(compressed_video,'rb').read()
            data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
            video_output.append_display_data(HTML("""<video width="80%%" height=400 controls><source src="%s" type="video/mp4"></video>""" % data_url))
            

    btn_play.on_click(run_video)
    data_box = widgets.HBox([label_data_files, ddl_data_selected])
    data_btn_box = widgets.VBox([widgets.HBox([btn_play]),video_output])
    display(data_box)
    display(data_btn_box)

    return

def load_STEP():
    %cd /content/gdrive/MyDrive/NVIDIA_STEP/STEP

    ! pip install torch==1.6.0+cu101 torchvision==0.7.0+cu101 -f https://download.pytorch.org/whl/torch_stable.html -q
    # !pip install ffmpeg-python -q
    ! python setup.py build develop -q

    import os
    import sys
    # import TSU.json_util as json_util
    # from TSU.json_util import test_train_json


    import shutil
    import torch
    ##Turn off warnings for python
    import warnings
    warnings.filterwarnings('ignore')
    warnings.simplefilter('ignore')
    warnings.filterwarnings("ignore", category=DeprecationWarning) 

    # Widget
    btn_check_if_cuda = widgets.Button(description="Check if your device is CUDA supported",
        layout = btn_layout, button_style='info')

    cuda_output = widgets.Output()

    def check_if_cuda(b):
        with cuda_output:
            cuda_output.clear_output()
            if (torch.cuda.is_available()):
                print('Your computer is CUDA supported.')
            else:
                print('Your computer is not CUDA supported.')
        # --- write script here to check if user's machine if cuda supported
        return True

    btn_check_if_cuda.on_click(check_if_cuda)

    # Display
    display(btn_check_if_cuda, cuda_output)

    display(Markdown("# ***NVIDIA STEP Collab Version***"))
    display(Markdown("## Extraction Code"))
    display(Markdown("* Ensure that frames extraction is performed before performing inferencing."))
    display(Markdown("* Issue No: https://ict3104-team03-2022.atlassian.net/browse/T03-64"))
    def populateList(fileDirectory, fileType=""):
        folder_files = os.listdir(fileDirectory) #You can also use full path.
        fileList = []
        for file in folder_files:
            if file.endswith(fileType) or fileType == "":
                fileList.append(file)
        return fileList

    fileListSTEP = populateList('/content/drive/MyDrive/NVIDIA_STEP/STEP/datasets/demo/', ".mp4")
    # Widget
    label_vid_files= widgets.Label("Select Video(s):", layout = label_layout)
    ddl_data_selected = widgets.Dropdown(
        options = fileListSTEP,
        value= fileListSTEP[0],
        layout = ddl_layout)

    btn_extract = widgets.Button(description='Extract Frames', layout = btn_sm_layout, button_style='info')
    
    # Display output
    extract_output = widgets.Output()
    
    def extract_video(b):
        with extract_output:
            extract_output.clear_output()
            # run extraction
            %run /content/drive/MyDrive/NVIDIA_STEP/STEPscripts/extract_frames.py -video_name $ddl_data_selected.value

    btn_extract.on_click(extract_video)

    # Display
    data_box = widgets.VBox([widgets.HBox([label_vid_files, ddl_data_selected])])
    data_btn_box = widgets.VBox([widgets.HBox([btn_extract]),extract_output])

    display(data_box)
    display(data_btn_box)

    display(Markdown("***"))
    display(Markdown("## Run Inference Code"))

    btn_infer = widgets.Button(
        description='Run Inferencing', layout = btn_sm_layout, button_style='info'
    )

    inference_output = widgets.Output()
    def run_inference(b):
        with inference_output:
            inference_output.clear_output()
            print("running inferencing")
            %run /content/drive/MyDrive/NVIDIA_STEP/STEP/demo.py

    btn_infer.on_click(run_inference)
    data_btn_box = widgets.VBox([widgets.HBox([btn_infer]),inference_output])
    display(data_btn_box)

    display(Markdown("***"))
    display(Markdown("## Run Inference Video"))

    import cv2
    import glob
    from base64 import b64encode
    from IPython.display import HTML
    ## UI to run inferencing

    btn_play = widgets.Button(
        description='Play video', layout = btn_sm_layout, button_style='info'
    )

    video_output = widgets.Output()

    def run_video(b):
        with video_output:
            video_output.clear_output()
            %cd /content/drive/MyDrive/NVIDIA_STEP/STEP
            print("Generating video...")
            mp4 = open('/content/drive/MyDrive/NVIDIA_STEP/STEP/datasets/demo/frames/results/P02T02C06/output.mp4','rb').read()
            data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
            video_output.append_display_data(HTML("""<video width=600 controls><source src="%s" type="video/mp4"></video>""" % data_url))

            # frame_path = "./datasets/demo/frames/results/{}/*.jpg".format(ddl_data_selected.value.replace(".mp4", ""))
            # output_path = './datasets/demo/frames/results/{}/output.mp4'.format(ddl_data_selected.value.replace(".mp4",""))
            # print("generating video")
            # # !ffmpeg -framerate 30 -pattern_type glob -i $frame_path -c:v libx264 -pix_fmt yuv420p $output_path -hide_banner -loglevel error
            # !ffmpeg -framerate 30 -pattern_type glob -i "./datasets/demo/frames/results/P02T02C06/*.jpg" -c:v libx264 -pix_fmt yuv420p './datasets/demo/frames/results/P02T02C06/output.mp4' -hide_banner -loglevel error

            # 
            # print("Video has been generated.")
            

    btn_play.on_click(run_video)
    data_btn_box = widgets.VBox([widgets.HBox([btn_play]),video_output])
    display(data_btn_box)

    display(Markdown("***"))
    display(Markdown("## Training"))
    display(Markdown("Steps taken to perform training:"))
    display(Markdown("1.   Convert csv file to pkl for both train.csv and val.csv"))
    display(Markdown("2.   Perform frame extraction for the videos to be trained and evaluated"))
    display(Markdown("3.   Input parameters for training "))
    display(Markdown("4.   Run training"))
    display(Markdown("note: Ensure that you have cuda enabled in order to perform training. Training will be conducted using the classification pre-trained weights (mAP = 26.4%). "))
    display(Markdown("Issue No: https://ict3104-team03-2022.atlassian.net/browse/T03-32"))
    display(Markdown("#### 3.1 Convert csv to pkl"))
    display(Markdown("note: Ensure that u have downloaded ava_train_v2.1.csv and ava_val_v2.1.csv from the AVA dataset. "))
    display(Markdown("[Link to download](https://research.google.com/ava/)"))
    
    ## Transform the origional annotation files in csv format to pickle files
    btn_transform = widgets.Button(
        description='Generate pickle files', layout = btn_sm_layout, button_style='info')
    # Display output
    pkl_output = widgets.Output()
    def extract_pkl(b):
        with pkl_output:
            # Create Video File
            pkl_output.clear_output()
            %cd /content/drive/MyDrive/NVIDIA_STEP/STEP
            %run scripts/generate_label.py datasets/ava/label/ava_train_v2.1.csv
            %run scripts/generate_label.py datasets/ava/label/ava_val_v2.1.csv

    btn_transform.on_click(extract_pkl)
    data_btn_box = widgets.VBox([widgets.HBox([btn_transform]),pkl_output])
    display(data_btn_box)

    display(Markdown("#### 3.2 Frame extraction for training videos"))
    display(Markdown("In order to perform training, testing and inference, videos need to be extracted into frames, where every 1 second, an image will be extracted."))
    display(Markdown("Ensure that you have downloaded the videos from the ava datasets, and the pkl extracted from 3.1 should match to the videos extracted here."))
    

    btn_extract = widgets.Button(
        description='Extract frames', layout = btn_sm_layout, button_style='info'
    )
    # Display output
    extract_output = widgets.Output()
    def extract_video(b):
        with extract_output:
            # Create Video File
            extract_output.clear_output()
            %run /content/drive/MyDrive/NVIDIA_STEP/STEP/scripts/extract_clips.py

    btn_extract.on_click(extract_video)
    data_btn_box = widgets.VBox([widgets.HBox([btn_extract]),extract_output])
    display(data_btn_box)

    display(Markdown("Run training"))

    # Specify a name for this new model using appropriate UI elements.
    desired_model_name = widgets.Text(
        placeholder='Type something..',
        disabled=False
    )
    # Provide user with epoch selection
    epochs = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        disabled=False
    )
    # Provide user with batch size
    user_batch_size = widgets.BoundedIntText(
        min=1,
        # max=1000, # to be defined and added as necessary
        step=1,
        disabled=False
    )
    train_button = widgets.Button(
        description='Train Model',
        disabled = False,
        button_style='info',
        layout=btn_layout
    )
    # Output train
    trainOutput = widgets.Output()
    # Functionality for user input
    def run_train(b):
        with trainOutput:
            clear_output()
            if desired_model_name.value == "" or desired_model_name.value == None:
                print("Model name is required.")
            elif epochs.value == "" or epochs.value == None:
                print("Batch size is required.")
            else:
                %cd /content/drive/MyDrive/NVIDIA_STEP/STEP/scripts
                !bash train_step.sh $user_batch_size.value $desired_model_name.value $epochs.value
    training_modelName_label = widgets.Label("Model Name", layout=label_layout)
    training_batchSize_label = widgets.Label("Batch Size", layout=label_layout)
    training_epochs_label = widgets.Label("Epoch", layout=label_layout)

    train_button.on_click(run_train)
    inputBox = widgets.VBox([widgets.HBox([training_modelName_label, desired_model_name]),
    widgets.HBox([training_batchSize_label, user_batch_size]),
    widgets.HBox([training_epochs_label, epochs])])
    display(inputBox)

    display(widgets.VBox([train_button,trainOutput]))

    display(Markdown("***"))
    display(Markdown("## 4. Testing"))
    display(Markdown("Steps to perform testing:"))
    display(Markdown("1.   Select a training folder"))
    display(Markdown("2.   Select a training model from the folder"))
    display(Markdown("3.   Run Testing"))
    display(Markdown("Issue No: https://ict3104-team03-2022.atlassian.net/browse/T03-65"))


    # TODO create ui to choose model to test

    def populateList(fileDirectory, fileType=""):
        folder_files = os.listdir(fileDirectory) #You can also use full path.
        fileList = []
        for file in folder_files:
            if file.endswith(fileType) or fileType == "":
                fileList.append(file)
        if (len(fileList) == 0):
            fileList.append("No Models found.")
        return fileList

    folderList = populateList('/content/drive/MyDrive/NVIDIA_STEP/STEP/datasets/ava/cache')
    # Widget to select folder
    label_folder= widgets.Label("Select Folder:", layout = label_layout)
    ddl_folder_selected = widgets.Dropdown(
        options = folderList,
        value= folderList[0],
        layout = ddl_layout)

    # onchange method
    def on_change(change):
          if change['type'] == 'change' and change['name'] == 'value':
              fileList = populateList('/content/drive/MyDrive/NVIDIA_STEP/STEP/datasets/ava/cache/' + ddl_folder_selected.value, ".pth")
              ddl_file_selected.options = fileList

    ddl_folder_selected.observe(on_change)

    # widget to select file from folder
    fileList = populateList('/content/drive/MyDrive/NVIDIA_STEP/STEP/datasets/ava/cache/' + ddl_folder_selected.value, ".pth")
    label_file= widgets.Label("Select file:", layout = label_layout)
    ddl_file_selected = widgets.Dropdown(
        options = fileList,
        value= fileList[0],
        layout = ddl_layout,
        disabled = False)

    btn_test = widgets.Button(
        description='Run Testing', layout = btn_sm_layout, button_style='info'
    )
    # Display output
    test_output = widgets.Output()
    def run_test(b):
        with test_output:
            test_output.clear_output()
            if (ddl_file_selected.value == "No Models found."):
                print("Please select a valid model.")
                return
            %run /content/drive/MyDrive/NVIDIA_STEP/STEP/test.py -model $ddl_file_selected.value -folder $ddl_folder_selected.value
    btn_test.on_click(run_test)

    # Display
    folder_box = widgets.VBox([widgets.HBox([label_folder, ddl_folder_selected])])
    file_box = widgets.VBox([widgets.HBox([label_file, ddl_file_selected])])
    data_btn_box = widgets.VBox([widgets.HBox([btn_test]),test_output])

    display(folder_box)
    display(file_box)
    display(data_btn_box)

    return
```
